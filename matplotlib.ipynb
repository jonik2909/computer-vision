```python
import matplotlib.pyplot as plt
import numpy as np
```

# Basic Line Plot

```python
x = np.array([1,2,3,4])
y = np.array([10,20,25,30])

plt.plot(x,y)
plt.title("Basic Line Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# Lines Styles and Colors

```python
x = np.array([1,2,3,4])
y = np.array([3,8,1,10])

plt.plot(y, linestyle='dotted', color='green', marker='o')
plt.title("Styled Line Example")
plt.show()
```

# Multiple Lines in One Graph with Numpy and Matplotlib

```python
x = np.array([1,2,3,4])
y1 = np.array([2,4,6,8])
y2 = np.array([1,3,5,7])

plt.plot(x, y1, label="Line 1", color="red", marker="s")
plt.plot(x, y2, label="Line 2", color="blue", marker="o", linestyle="--")
plt.title("Multiple Lines Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.legend()
plt.show()
```

# Bar Chart with Numpy and Matplotlib

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([5, 9, 1, 7])

plt.bar(x, y, color="orange")
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

# Scatter Plot with Numpy and Matplotlib

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 4, 6, 3, 7])

plt.scatter(x, y,  color="purple", s=100)
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# Histogram with Numpy and Matplotlib

```python
data = np.array([3, 4, 5, 6, 8, 8, 8, 9, 10])

plt.hist(data, bins=5, color="skyblue", edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Count")
plt.show()
```

# Subplots(Multiple Graphs Together)

```python
x = np.array([1, 2, 3, 4])
y1 = np.array([2, 4, 6, 8])
y2 = np.array([3, 6, 9, 12])

plt.figure(figsize=(8, 4)) # width | height

plt.subplot(1, 2, 1) #
plt.plot(x, y1, color="red")
plt.title("Plot 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2, color="green")
plt.title("Plot 2")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
```
