```python
import cv2
```

```python
from google.colab.patches import cv2_imshow

image = cv2.imread("/content/cat_image.jpg")
# cv2.imshow("CAT", image) // for local

cv2_imshow(image) # for colab
```

```python
print("SHAPE:", image.shape)
# 463 => height
# 703 => width
# 3 => color channels (RGB) => OPENCV (BGR) => BLUE, GREEN, RED

print("HEIGHT:", image.shape[0])
print("WIDTH:", image.shape[1])
print("CHANNELS:", image.shape[2])
```

```python
(b, g, r) = image[100, 50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray, (300, 200))
cv2_imshow(resized)
```

```python
cropped = image[100:300, 150:350] # y, x
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_90_CLOCKWISE)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("flipped.jpg", flipped)
```
